```python
import pandas as pd

# Read the CSV file into a DataFrame
zipcode_df = pd.read_csv('/Users/shebanaderzad/Downloads/simplemaps_uszips_basicv1/uszips.csv')
zipcode_df.query('zip==1440')
```

```python
states = set(zipcode_df['state_id'])
state_abbreviations = [
    "AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DE", "FL", "GA",
    "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD",
    "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ",
    "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC",
    "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"
]
# Find state abbreviations that are additional and need to be removed
non_us_states = [val for val in states if val not in state_abbreviations]
non_us_states

# Remove additional abbreviations from the original list (if needed)
for val in non_us_states:
    states.remove(val)
```

```python
# Filter rows based on state abbreviations and put zips in list
zip_data_us = zipcode_df[zipcode_df['state_id'].isin(states)]
zip_codes_us = zip_data_us['zip'].tolist()
zip_codes_us
```

```python
import requests
from bs4 import BeautifulSoup

for zip_code in zip_codes_us:
    # Construct the URL based on the zip code
    url = f"https://crimegrade.org/safest-places-in-{zip_code}/"

    # Send a GET request to the URL
    response = requests.get(url)

    if response.status_code == 200:
        # Parse the HTML content with Beautiful Soup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract the crime data grade
        crime_data_grade = soup.find('div', class_='one_half').text
        print(f"Zip Code {zip_code}: {crime_data_grade}")
    else:
        print(f"Error accessing data for Zip Code {zip_code}")
```
